```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Data.csv')
data
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.fillna('0')
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.grid_search import GridSearchCV
```

```python
y = data['Target']
x = data.drop(['Target'],axis=1)
x = pd.get_dummies(x)
x.shape
```

```python
logit = LogisticRegression()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
logit.fit(xtrain,ytrain)
```

```python
pred1 = logit.predict(xtest)
```

```python
print(accuracy_score(pred1,ytest))
```

```python
dectree = DecisionTreeClassifier()
```

```python
dectree.fit(xtrain,ytrain)
```

```python
pred2 = dectree.predict(xtest)
print(accuracy_score(pred2,ytest))
```

```python
randfor = RandomForestClassifier(criterion='gini',max_leaf_nodes=30)
```

```python
randfor.fit(xtrain,ytrain)
```

```python
pred3 = randfor.predict(xtest)
print(accuracy_score(pred3,ytest))
```

```python
param_test = {'n_estimators':[125,150,175]}
gridsearch = GridSearchCV(estimator=randfor,param_grid=param_test,verbose = 10)
gridsearch.fit(xtrain,ytrain)
```

```python
pred4 = gridsearch.predict(xtest)
print(accuracy_score(pred4,ytest))
```

```python
adaboost = AdaBoostClassifier(base_estimator=randfor,n_estimators=175)
```

```python
adaboost.fit(xtrain,ytrain)
```

```python
pred5 = adaboost.predict(xtest)
print(accuracy_score(pred5,ytest))
```

```python
mpl = MLPClassifier()
```

```python
mpl.fit(xtrain,ytrain)
```

```python
pred6 = mpl.predict(xtest)
```

```python
print(accuracy_score(pred6,ytest))
```

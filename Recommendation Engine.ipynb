```python
book1 = pd.read_csv('book_recom.csv')
book1.head()
```

```python
book1.isnull().sum()
```

```python
book_ratings = book1[['userID','bookRating','bookTitle']]
book_ratings.head()
```

```python
book_pivot_table = book_ratings.pivot_table(columns='userID',values='bookRating',index='bookTitle').fillna(0)
book_pivot_table
```

```python
book_matrix = csr_matrix(book_pivot_table.values)
print(book_matrix)
```

```python
model_knn = NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(book_matrix)
```

```python
query_index1 = 0
distance,indices = model_knn.kneighbors(book_pivot_table.iloc[query_index1, :].values.reshape(1,-1),n_neighbors=6)
print(distance,indices)
```

```python
book_selection = book_pivot_table.index
book_selection = book_selection.unique()
book_selection
```

```python
books = pd.DataFrame(book_selection)
books
```

```python
books[indices]
```
